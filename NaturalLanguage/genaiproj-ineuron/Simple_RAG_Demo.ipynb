````python
system_prompt = """
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""

query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
````

````python
import torch
````

````python
llm = HuggingFaceLLM(
    context_window = 4096,
    max_new_tokens = 256,
    generate_kwargs = {"temperature": 0.0, "do_sample": False},
    system_prompt = system_prompt,
    query_wrapper_prompt= query_wrapper_prompt,
    tokenizer_name = "mistralai/Mistral-7B-Instruct-v0.1",
    model_name = "mistralai/Mistral-7B-Instruct-v0.1",
    device_map = 'auto',
    # uncomment this if using CUDA to reduce memory usage
    # model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True} ## 使用此行，需要使用GPU才行 (功能quantization)
)
````

````python
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings ## 安裝 langchain-community
from llama_index import ServiceContext
from llama_index.embeddings import LangchainEmbedding
````

````python
emded_model = LangchainEmbedding(
  HuggingFaceBgeEmbeddings(model_name ="sentence-transformers/all-mpnet-base-v2")
  ## 安裝 sentence-transformers
)
````

````python
service_context = ServiceContext.from_defaults(
    chunk_size = 1024,
    llm = llm,
    embed_model = emded_model
)
````

````python
index = VectorStoreIndex.from_documents(documents, service_context= service_context)
````

<font color="red">如果沒使用service_context，此為默認用法，使用openai接口，需要設置OPENAI_API_KEY。
```python

```
<font color="red">將`index`保留在本地端，執行後會生成 `/storage` 資料夾，裡面包含 5個 json 檔
```python
index.storage_context.persist()
```

````python
query_engine = index.as_query_engine()
````

````python
# 使用 CPU 將耗時 (1hr47m)，除了CPU速度較慢，CPU不能使用quantize的模型，表示無法節省時間！
query = "What is Positive and Negative Correlation?"
response = query_engine.query(query)
print(response)
````

````python
# 使用 GPU 耗時 35 秒 (quantization) 或 28 分鐘 (no quantization)
query = "What is Positive and Negative Correlation?"
response = query_engine.query(query)
print(response)
````
