## give a function that can sort the given list

````python
my_list = [64, 34, 25, 12, 22, 11, 90]
````

````python
def bubble_sort(my_lst):
    n=len(my_lst)

    for i in range(n):
        for j in range(n-i-1):
           if my_lst[j]>my_lst[j+1]:
              my_lst[j], my_lst[j+1]=my_lst[j+1], my_lst[j]
    return my_lst
````

````python
print(f"my original list {my_list}")
my_lst = bubble_sort(my_list)
print(f"my sorted list {my_lst}")
````

## time complexity of the above code

O(n^2)

# 8. Minimum value in dict

my_dict={'a':10, 'b':5, 'c':8, 'd':3}

````python
my_dict={'a':10, 'b':5, 'c':8, 'd':3}
````

````python
min(my_dict.values())
````

````python
min_key=None
min_value=float('inf')
````

````python
min_value
````

````python
for key, value in my_dict.items():
    if value < min_value:
        min_value = value
    if key < min_key:
        min_key = key

print(min_value)
print(min_key)
````

# 12. factorial with recursion

````python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
````

````python
factorial(0), factorial(1), factorial(2), factorial(5)
````

````python
factorial(50)
````

# 13. fabonacci series

````python
n=10
num1 = 0
num2 = 1

next_num = num2
count = 1
````

````python
while count<=n:
    print(next_num, end=" ")

    count=count+1
    num1, num2 = num2, next_num
    next_num=num1+num2

print()
````
