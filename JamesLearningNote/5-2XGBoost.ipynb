# 實作
<font color="red">首先載入鐵達尼號的資料集，簡單使用鐵達尼資料中的性別、艙等、年齡來預測乘客是否存活

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
titanic = pd.read_csv('Titanic/train.csv')
```

```python
titanic
```

<font color="red">選取性別、艙等、年齡、是否存活以及將年齡補空值，並將資料拆成Train以及Test

```python
X = titanic[['Pclass', 'Age', 'Sex']]
y = titanic['Survived']
```

```python
X['Age'] = X['Age'].fillna(X['Age'].mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)
```

```python
X_train
```

<font color="red">將性別轉為0,1的數值型態

```python
X_train['Sex'] = X_train['Sex'].map({'male':0,'female':1})
X_test['Sex'] = X_test['Sex'].map({'male':0,'female':1})
```

```python
X_train
```

<font color="red">什麼參數都沒調整的情況下RandomForest的正確率是81.6%? (實測卻是0.8385)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
print('The accuracy of Random Forest Classifier on testing set:', rfc.score(X_test, y_test))
```

<font color="red">而在什麼參數都沒調整的情況下XGBoost的正確率是83.8%，在什麼事都沒做的情況下(只是換Model)就高出了2.2%，相當划算的投資

```python
from xgboost import XGBClassifier
```

```python
xgbc = XGBClassifier()
```

```python
xgbc.fit(X_train, y_train)
```

```python
print('The accuracy of eXtreme Gradient Boosting Classifier on testing set:', xgbc.score(X_test, y_test))
```
