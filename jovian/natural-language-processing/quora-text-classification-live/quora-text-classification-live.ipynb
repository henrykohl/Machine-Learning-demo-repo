### Bag of Words Intuition

1. Create a list of all the words across all the text documents <font color="red">建立list(把所有句子)</font>
2. You convert each document into vector counts of each word <font color="red">建立vector(對每一個句子)</font>


Limitations:
1. There may be too many words in the dataset
2. Some words may occur too frequently
3. Some words may occur very rarely or only once
4. A single word may have many forms (go, gone, going or bird vs. birds)

### Configure Count Vectorizer Parameters

````python
# Lecture Video
# def tokenize(text):
#   return [word for word in word_tokenize(text) if word.lower() not in english_stopwords]
````

````python
# tokenize('What is the really (dealing) here?')
````

````python
stemmer = SnowballStemmer(language='english') # 前面已經建立過 ， 這裡就不需要再次執行
````

````python
def tokenize(text):
    return [stemmer.stem(word) for word in word_tokenize(text)]
````

````python
tokenize('What is the really (dealing) here?')
````

````python
vectorizer = CountVectorizer(lowercase=True,        # lowercase
                tokenizer=tokenize,      # tokenize 函數，包含 word_tokenize 與stemmer.stem 兩個功能
                stop_words=english_stopwords, # stop_words
                max_features=1000)
````

````python
%%time
vectorizer.fit(sample_df.question_text)
````

````python
len(vectorizer.vocabulary_)
````

````python
vectorizer.get_feature_names_out()[:100] # 因為 max_features=1000
````

````python
%%time
inputs = vectorizer.transform(sample_df.question_text)
````

````python
inputs.shape
````

````python
sample_df.question_text.values[0]
````

## ML Models for Text Classification

Outline:

- Create a training & validation set
- Train a logistic regression model
- Make predictions on training, validation & test data

### Split into Training and Validation Set

````python
sample_df
````

````python
inputs.shape
````

````python
from sklearn.model_selection import train_test_split
````

````python
train_inputs, val_inputs, train_targets, val_targets = train_test_split(inputs, sample_df.target, test_size=0.3, random_state=42)
````

````python
train_inputs.shape
````

````python
train_targets.shape
````

````python
val_inputs.shape
````

````python
val_targets.shape
````

### Train Logistic Regression model

````python
from sklearn.linear_model import LogisticRegression
````

````python
MAX_ITER = 1000
````

````python
model = LogisticRegression(max_iter=MAX_ITER, solver='sag') # 如果不設max_iter，fit過程會因為iteration次數出錯，因為資料太大
````

````python
%%time
model.fit(train_inputs, train_targets)
````

### Make predictions using the model

````python
train_preds = model.predict(train_inputs)
````

````python
train_targets # Series 類型
````

````python
train_preds # numpy.nparray 類型
````

````python
pd.Series(train_preds).value_counts()
````

````python
pd.Series(train_targets).value_counts() # 直接用 train_targets.value_counts() 既可
````
